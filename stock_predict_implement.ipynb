```python
import pandas as pd
samsung_stock_prices = pd.read_csv('samsung_2022-01-01~2023-02-02.csv')
samsung_stock_prices
```

```python
# '일자' 열의 날짜 형식 변경 ('20230102' -> '2023-01-02')
df['일자'] = pd.to_datetime(df['일자'], format='%Y%m%d').dt.strftime('%Y-%m-%d')
```

```python
# '일자' 열의 날짜 형식 변경 ('2023-01-02' 형식으로)
df['일자'] = pd.to_datetime(df['일자']).dt.strftime('%Y-%m-%d')

# '날짜' 열의 날짜 형식 변경 ('2023-01-02' 형식으로)
samsung_stock_prices['날짜'] = pd.to_datetime(samsung_stock_prices['날짜']).dt.strftime('%Y-%m-%d')



# 변경된 DataFrame 확인
print(df['일자'].head())
print(samsung_stock_prices['날짜'].head())
```

```python
# 두 DataFrame을 '일자' 및 '날짜'를 기준으로 조인
merged_df = pd.merge(df, samsung_stock_prices, how='left', left_on='일자', right_on='날짜')

# 조인한 결과에서 '시가' 컬럼을 원래 DataFrame에 추가
df['변동 %'] = merged_df['변동 %']
```
